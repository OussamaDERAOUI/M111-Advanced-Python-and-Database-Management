# Matplotlib exercices

In this exercise, we are using Pandas and Matplotlib to visualize Company Sales Data. The data is located under the folder `../data/company_sales_data.csv`.

![company_sales_data](../data/company_sales_data.webp)

```python
import matplotlib.pyplot as plt

import pandas as pd

data = pd.read_csv('~/Desktop/M111-Advanced-Python-and-Database-Management/day01/data/company_sales_data.csv')
data
```

## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:

The line plot graph should look like this.

**Expected output**

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>


```python
y = data.loc[:,'total_profit']
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)
rnge = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(y, rnge,label = 'profit data')
plt.legend(loc ='upper left')
plt.xlabel('profit range in dollar')
plt.ylabel('actual profit')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
piedata   = [data['facecream'].sum(), data['facewash'].sum(), 
             data['toothpaste'].sum(), data['bathingsoap'].sum(),
             data['shampoo'].sum(), data['moisturizer'].sum()]

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)
plt.pie(piedata, labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.legend(loc= "lower right")
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
fc  =data.loc[:,'facecream']
bs   = data.loc[:,'bathingsoap']
months = data.loc[:,'month_number']


s,subplt = plt.subplots(2, sharex=True)
subplt[0].plot(months, bs, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
subplt[0].set_title('Sales data of  a Bathingsoap')
subplt[1].plot(months, fw, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
subplt[1].set_title('Sales data of  a facewash')


plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
fc  =data.loc[:,'facecream']
fw   = data.loc[:,'facewash']
tp = data.loc[:,'toothpaste']
bs   = data.loc[:,'bathingsoap']
ss   = data.loc[:,'shampoo']
ms = data.loc[:,'moisturizer']
months = data.loc[:,'month_number']

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(6)

plt.plot([],[],color='m', label='face Cream')
plt.plot([],[],color='c', label='Face wash')
plt.plot([],[],color='r', label='Tooth paste')
plt.plot([],[],color='k', label='Bathing soap')
plt.plot([],[],color='g', label='Shampoo')
plt.plot([],[],color='y', label='Moisturizer')

plt.stackplot(months, fc, fw, tp, bs, ss, ms, colors=['m','c','r','k','g','y'])
plt.legend(loc="upper left")
plt.xlabel('Month Number')
plt.ylabel('Sale units')
plt.show()
```
